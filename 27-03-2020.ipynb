```python
x=np.arange(-9,10)
y=x**2
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("parabola")
plt.xticks(x)  #(np.arange(-9,10,4),to repressent some ticekes/points)
plt.yticks(y) #represents the y actual values
```

```python
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("parabola")
plt.xlim(-5,5) #Representing the graph
plt.ylim(10,50)
```

```python
x=np.arange(5,10)
plt.subplot(2,2,1,frameon=False)
plt.plot(x)
plt.title("x")
plt.subplot(2,2,2,projection='polar') # 2,2 subplot,first position
plt.plot(x,x+3)
plt.title("x+3")
plt.subplot(2,2,3)
plt.plot(x,x+x)
plt.title("x+x")
plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title('log x')
```
